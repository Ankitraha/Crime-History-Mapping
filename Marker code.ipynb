```python
import pandas as pd

# load the CSV file
df = pd.read_csv('crimes_colours_final.csv')

# create a dictionary to map integer codes to crime names
crime_map = {
    1: 'murder',
    2: 'sexual assault',
    3: 'aggravated assault',
    4: 'kidnapping',
    5: 'robbery'
}

# create a new column with the corresponding crime name
df['crime_name'] = df['crime_type'].map(crime_map)

# save the updated dataframe to a new CSV file
df.to_csv('crime_data_with_names1.csv', index=False)
```

```python
import folium
import pandas as pd

# Read in the dataset
df = pd.read_csv('crime_data_with_names.csv')

# Define the map center
center = [df['Latitude'].mean(), df['Longitude'].mean()]

# Create the map object
m = folium.Map(location=center, zoom_start=14)

# Define a color scheme for the different types of crimes
color_scheme = {
    1: 'red',
    2: 'orange',
    3: 'blue',
}

# Add a marker for each crime
for _, row in df.iterrows():
    try:
        color = color_scheme[row['crime_type']]
    except KeyError:
        color = 'gray'
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        icon=folium.Icon(color=color),
        popup=row['crime_name']  # Add a popup with the name of the crime
    ).add_to(m)

# Show the map
m
```

```python
import pandas as pd
import numpy as np
from scipy.spatial import ConvexHull

crime_data = pd.read_csv('crimes_colours_final.csv')

# Get the unique neighborhood IDs
neighborhood_ids = crime_data['Place'].unique()

# Create a folium map centered on the first data point
m = folium.Map(location=[crime_data['Latitude'][0], crime_data['Longitude'][0]], zoom_start=12)

# Loop over the neighborhood IDs
for neighborhood_id in neighborhood_ids:
    neighborhood_data = crime_data[crime_data['Place'] == neighborhood_id]
    coords = neighborhood_data[['Longitude', 'Latitude']].values
    if len(coords) > 0 and not np.isnan(coords).any():
        hull = ConvexHull(coords)
        polygon_coords = coords[hull.vertices]
        polygon = {
            "type": "Feature",
            "geometry": {
                "type": "Polygon",
                "coordinates": [polygon_coords.tolist()]
            },
            "properties": {}
        }
        folium.GeoJson(data=polygon, style_function=lambda x: {'fillColor': '#ff0000'}).add_to(m)

        
# Read in the dataset
df = pd.read_csv('crime_data_with_names1.csv')



# Define a color scheme for the different types of crimes
color_scheme = {
    1: 'red',
    2: 'orange',
    3: 'blue',
}

# Add a marker for each crime
for _, row in df.iterrows():
    try:
        color = color_scheme[row['crime_type']]
    except KeyError:
        color = 'gray'
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        icon=folium.Icon(color=color),
        popup=row['crime_name']  # Add a popup with the name of the crime
    ).add_to(m)
# Display the map
m
```
